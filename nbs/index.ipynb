## Compute a diff

## Full pipeline (no LLM)

````python
await process_diff(
    "The old cat sat on the mat.",
    "The new cat lay on the mat.",
    semantic_grouping=False
)
````

## Track and accept changes

````python
doc = DocumentState()
doc.add_segments([
    {"id": "1", "text": "The ", "segment_type": "keep", "explanation": "", "accepted": True},
    {"id": "2", "text": "old ", "segment_type": "remove", "explanation": "replacing", "accepted": False},
    {"id": "3", "text": "new ", "segment_type": "add", "explanation": "replacing", "accepted": False},
    {"id": "4", "text": "cat.", "segment_type": "keep", "explanation": "", "accepted": True},
])

print("before:", doc.get_working_text())
doc.accept_segment("2", "add")  # accept the new text
print("after: ", doc.get_working_text())
````
